```python
# 우리나라의 연간 1인당 국민소득을 각각 years, gdp에 저장
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010] 
gdp = [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3]
```

```python
import matplotlib.pyplot as plt
```

```python
 
plt.plot(years,gdp,color="r",marker="o",linestyle="dashed",label="gdp of korea") # 앞이 x축 값, 뒤가 y축 값 이다.

plt.title("GDP")
plt.ylabel("dollors")
plt.xlabel("years")
plt.legend()


plt.savefig("gdp_per_capita.png", dpi=150) # 그래프 저장
plt.show()# 출력되는 문장이 안나옴
```

```python
import numpy as np
```

```python
x=np.arange(0,10,0.1)
y=2*x

plt.plot(x,y,"r--")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim([-5,15])
plt.ylim([-5,30])
plt.grid()
plt.show()
```

```python
rand = np.random.randn(1000)
plt.figure(figsize=(10,5))
plt.plot(range(1000),rand,"ro:")
plt.title("numbers")
plt.show()
```

```python
x=np.linspace(-20,20,41)
y1=2*x
y2=x**2+5
y3=-x**2-5
plt.plot(x,y1,"r--", label="y=2x")
plt.plot(x,y2,"g^-", label="y=x**2+5") #label 이 보이려면 plt.legend() 함수를 써줘야한다.
plt.plot(x,y3,"b*:", label="y=-x^2-5")

plt.legend()
plt.grid()
plt.xlim([-30,30])
plt.ylim([-30,30])
plt.show()
```
